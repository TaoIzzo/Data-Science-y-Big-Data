```
def parser(line):
    fields = line.split(',')
    nline = fields[0]
    nid = fields[1]
    gender = fields[2]
    customer_type = fields[3]
    age = int(fields[4])
    type_of_travel = fields[5]
    class_n = fields[6]
    flight_distance = int(fields[7])
    inflight_wifi_service = fields[8]
    departure_arrival_time_convenient = fields[9]
    ease_of_online_booking = int(fields[10])
    gate_location = fields[11]
    food_and_drink = fields[12]
    online_boarding = fields[13]
    seat_comfort = fields[14]
    inflight_entertainment = fields[15]
    on_board_service = fields[16]
    leg_room_service = fields[17]
    baggage_handling = fields[18]
    checkin_service = fields[19]
    inflight_service = fields[20]
    cleanliness = fields[21]
    departure_delay_in_minutes = int(fields[22])
    arrival_delay_in_minutes = fields[23]
    satisfaction = fields[24]
```
